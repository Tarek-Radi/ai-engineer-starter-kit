```python
import gradio as gr

def greet(name, intensity):
    return "Hello, " + name + "!" * int(intensity)

demo = gr.Interface(
    fn=greet,
    inputs=["text", "slider"],
    outputs=["text"],
    api_name="predict"
)

demo.launch()
```

```python
def chat(message, history):
    return "You said: " + message

demo = gr.ChatInterface(fn=chat)
demo.launch()
```
